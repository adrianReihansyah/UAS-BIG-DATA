```r
data$Product_holding <- as.factor(data$Product_holding)
set.seed(600)
```

```r
ind <- sample(2, nrow(data), replace = TRUE, prob = c(0.7, 0.3))
```

```r
training <- data[ind==1,]
```

```r
testing <- data[ind==2,]
```

```r
library(nnet)
```

```r
training$Product_holding <- relevel(training$Product_holding, ref="1")
```

```r
bil <- multinom(Product_holding~., data = training)
```

```r
summary(bil)
```

```r
z <- summary(bil)$coefficients/summary(bil)$standard.errors
```

```r
p <- (1 - pnorm(abs(z), 0, 1)) * 2
```

```r
p
```

```r
p <- predict(bil, training)
```
